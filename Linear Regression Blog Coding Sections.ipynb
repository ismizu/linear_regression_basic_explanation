## Linear Graphs

```python
import numpy
import matplotlib.pyplot as plt
```

```python
#Create x, y coordinates with a strong linear relationship
x = np.arange(20)
y = 3*x + 5

#Create the line of best fit
a, b = np.polyfit(x, y, 1)


fig, ax = plt.subplots()
#Plot the initial x and y values
ax.scatter(x, y)
#Plot the line of best fit
ax.plot(x, a*x+b, c = 'red')

ax.set_title('Price of Apples')
ax.set_xlabel('Apple Types')
ax.set_ylabel('Apple Prices')
ax.legend(['Line of Best Fit', 'Apple Prices']);
```
